### ***<span style= "color:#85BB65;"> Using mean, median and mode </span>***

```python
# Imputation using mean, median and mode
# Always remember mean and median are used for numerical data and mode is used for categorical data.
# Mean effects with the outliers and median is robust to outliers.

# Impute missing value in 'age' column with mean using pandas
# df['age'] = df['age'].fillna(df['age'].mean())

# Impute missing values in 'age' column with median using pandas
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Due to too much missing values in deck and its also not that important so we drop this column
df.drop('deck', axis=1, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Now as you can see embarked and embarked_town are catagorical columns so we have to use Mode here

# Impute missing values in embarked and embarked_town usig mode 

# first check the mode
df[['embark_town', 'embarked']].value_counts()
df[['embark_town', 'embarked']].mode()

# # now we replace with it
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

### ***<span style= "color:#85BB65;"> Using sk_learn </span>***

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.impute import SimpleImputer
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Impute age column using simpleimputer from sklearn
imputer = SimpleImputer(strategy='median')
df['age'] = imputer.fit_transform(df[['age']])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# multivariate imputation
df = sns.load_dataset('titanic')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
df['age'] = imputer.fit_transform(df[['age']])
df.isnull().sum().sort_values(ascending=False)
```
