## ***<span style= "color:purple;">Imports </span>***

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
```

```python
df = sns.load_dataset('titanic')
```

```python
age = df[['age']]
```

### ***<span style= "color:#85BB65;"> Standard scaling</span>***

```python
# Standard scaler
stand_scaler = StandardScaler()
scaled_age = stand_scaler.fit_transform(age)

scaled_age_df = pd.DataFrame(scaled_age, columns=['age'])
scaled_age_df.head()
```

### ***<span style= "color:#85BB65;"> Min Max scaling</span>***

```python
# Min Max scaler
MinMax_scaler = MinMaxScaler()
MinMaxScaled_age = MinMax_scaler.fit_transform(age)

MinMax_Scaled_age_df = pd.DataFrame(MinMaxScaled_age, columns=['age'])
MinMax_Scaled_age_df.head(10)
```
