### ***<span style= "color:#85BB65;">Label encoding</span>***

## ***<span style= "color:purple;">Imports </span>***

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = sns.load_dataset('tips')
```

```python
df['time'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum().sort_index(ascending=False)
```

### ***<span style= "color:#85BB65;">Label encoding</span>***

```python
time = df['time']

le = LabelEncoder()
encoded_time = le.fit_transform(time)
encoded_time_df = pd.DataFrame(encoded_time, columns=['encodded time'])
encoded_time_df['encodded time'].value_counts()
```

### ***<span style= "color:#85BB65;">One-hot encoding</span>***

```python
time = df[['time']]


ohe = OneHotEncoder()
encoded_time = ohe.fit_transform(time)
encoded_time_df = pd.DataFrame(encoded_time, columns=['encodded time'])
encoded_time_df.head()
```
