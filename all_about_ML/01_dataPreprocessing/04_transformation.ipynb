## ***<span style= "color:purple;">Imports </span>***

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import PowerTransformer
```

### ***<span style= "color:#85BB65;">Box-cox transformation</span>***

```python
np.random.seed(0)
expo_numbers = np.random.exponential(size=1000, scale=2)
df = pd.DataFrame(expo_numbers, columns=['values'])
df.head()
```

```python
sns.histplot(df['values'], kde=True)
```

```python
pt_boxcox = PowerTransformer(method='box-cox')

df['box_cox'] = pt_boxcox.fit_transform(df[['values']])
sns.histplot(df, kde=True)
```

### ***<span style= "color:#85BB65;">yo-johnson transformation</span>***

```python
pt_yo_johnson = PowerTransformer(method='yeo-johnson')

df['yo-jon'] = pt_yo_johnson.fit_transform(df[['values']])
sns.histplot(df[['values','yo-jon']], kde=True)
```
