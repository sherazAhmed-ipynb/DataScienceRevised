### ***<span style= "color:#56A5EC;"> Key terms </span>***
- **<span style= "color:#CD5C5C;"> Data points</span>** (data points in the data)
- **<span style= "color:#CD5C5C;"> X-axis / Y-axis</span>** (x-axis=>input of data like time | y-axis=> output of data like temperature)
- **<span style= "color:#CD5C5C;"> Figure</span>** (like whole window where multiple graphs can be plot )
- **<span style= "color:#CD5C5C;"> Axes</span>** (small rectagular draw on Figure)
- **<span style= "color:#CD5C5C;"> Plot</span>** (where data is displayed)
- **<span style= "color:#CD5C5C;"> Marker</span>** (A dot like [o,s,D,x,+])
- **<span style= "color:#CD5C5C;"> Line style</span>** (countinous connections b/w data points. style of line like [-,..., ___,+++])
- **<span style= "color:#CD5C5C;"> Color</span>** (color of data point, marker or line like [b,g,r or black, green or hexcode])
- **<span style= "color:#CD5C5C;"> Legend</span>** (a small box in graph which tells mean of dot, line with different colors)
- **<span style= "color:#CD5C5C;"> Label</span>** (donated name for x-axis and y-axis)
- **<span style= "color:#CD5C5C;"> Title</span>** (head line of graph)
- **<span style= "color:#CD5C5C;"> Grid</span>** (grid behind a graph)
- **<span style= "color:#CD5C5C;"> DPI</span>** (dots per inch | higher dpi higher resolution of graph which then been saved)

## ***<span style= "color:#EE9A4D;">Line graph </span>***

```python
# Starting wiht very basic line graph
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] # x-axis
cake_sales = [10,15,8,5,17,7,20] # y-axis

# fmt is used to define the format of the plot
# fmt = [color][marker][line]
# fmt = 'ro-' means red color, circle markers, solid line
#plt.plot(days, cake_sales, 'go--')

# You can customize the plot in various ways 
# (color = 'green', marker = 'x', marker size = 10, markeredgecolor = 'red', marheredgesize = 2, markerfacecolor = 'yellow', linestyle = '--', linewidth = 2, linecolor = 'blue',)
plt.plot(days, cake_sales, color = 'green', marker = 'x', markersize = 10, markeredgecolor = 'red', markeredgewidth = 2, markerfacecolor = 'yellow', linestyle = '--', linewidth = 2)
plt.title("Cake sales over a week")
plt.xlabel("Days")
plt.ylabel("Number of cakes sold")
#plt.yticks()
plt.legend(["cake sales"], loc = 'upper left')
plt.grid(True, linestyle = ':')
plt.show()
```

## ***<span style= "color:#EE9A4D;">Bar graph </span>***

```python
# Creating a bar graph
# Bar charts are used to represent categorical data.

products = ['A','B','C','D']
products_color = ['red', 'blue', 'green', 'orange']
sales = [250, 150,300,200]
plt.bar(products, sales, color = products_color, label = 'Sales 2025')
plt.title("Sales of products")
plt.xlabel("Products")
plt.ylabel("Sales in units")

plt.show()
```
